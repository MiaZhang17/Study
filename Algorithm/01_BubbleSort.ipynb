## Bubble Sort

- 長度為L的數列，檢查L-1次
- 每次從`index = 0` 檢查到倒數第二個index
    - 若目前index數字 > 下一個index的數字，2個數字交換位置(swap)
    - 再到下一個index繼續檢查

#### Bubble Sort 迴圈次數優化
- 第一回合檢查完後，最大的數字會被swap到最後一位 (最大數字已排序完成)
- 第二回合檢查完後，第二大的數字會被swap到倒數第二位 (最後2位置的數字皆已排序完成)
- 第三回合檢查完後，第三大的數字會被swap到倒數第三位 (最後3位置的數字皆已排序完成)

以此類推
- Round 0：必需檢查到Index：L-1 (全都尚未排序)
- Round 1：只需檢查到Index：L-2 (L-1以後的皆已排序完成)
- Round 2：只需檢查到Index：L-3 (L-2以後的皆已排序完成)

#### Time Complexity O(n) = $n^2$ 
- $(n-1)(n-1-i)$

```python
def swap(arr, index_1, index_2):
    temp = arr[index_1]
    arr[index_1] = arr[index_2]
    arr[index_2] = temp

def bubble_sort_unoptimized(arr):
    iteration_count = 0
    for i, el in enumerate(arr):
        for idx in range(len(arr) - 1):
            iteration_count += 1
            print('Check', arr[idx], arr[idx + 1], 'round:', i, ', index:', idx)
            if arr[idx] > arr[idx + 1]:
                swap(arr, idx, idx + 1)
                print('swap:', arr[idx+1], arr[idx], arr)

    print("PRE-OPTIMIZED ITERATION COUNT: {0}".format(iteration_count))

def bubble_sort(arr):
    iteration_count = 0
    for i in range(len(arr)):
        print(arr)
        # iterate through unplaced elements
        for idx in range(len(arr) - i - 1):
            iteration_count += 1
            print('Check', arr[idx], arr[idx + 1], 'round:',i,', index:', idx)
            if arr[idx] > arr[idx + 1]:
                # replacement for swap function
                arr[idx], arr[idx + 1] = arr[idx + 1], arr[idx]
                print('swap:', arr[idx+1], arr[idx], arr)
        
    print("POST-OPTIMIZED ITERATION COUNT: {0}".format(iteration_count))
```

```python
a = [5, 1, 8, 3]
print(a)
bubble_sort_unoptimized(a)
print("POST SORT: {0}".format(a))
```

```python
a = [5, 1, 8, 3]
print(a)
bubble_sort(a)
print("POST SORT: {0}".format(a))
```

```python
nums = [9, 8, 7, 6, 5, 4, 3, 2, 1]
print("PRE SORT: {0}".format(nums))

bubble_sort_unoptimized(nums.copy())
bubble_sort(nums)
print("POST SORT: {0}".format(nums))
```
